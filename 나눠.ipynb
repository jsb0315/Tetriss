```python
def find_common_divisors(a, b):
    common_divisors = []
    gcd_value = gcd(a, b)

    for i in range(1, gcd_value + 1):
        if gcd_value % i == 0:
            common_divisors.append(i)

    return common_divisors

def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

import cv2
import numpy as np

# 이미지를 로드합니다. 이 예시에서는 "input.jpg" 파일을 사용합니다.
image = cv2.imread("images/dog.jpg")

height, width, ch = image.shape

common_divisors = find_common_divisors(width, height)
print(f"{width}와 {height}의 모든 공약수: {common_divisors}")

for i in common_divisors:
    print(i, width/i, height/i)
```

```python
# [1, 2, 3, 4, 5, 6, 8, 10, 12, 15, 20, 24, 25, 30, 40, 50, 60, 75, 100, 120, 150, 200, 300, 600]
block = 30 # n등분(생성 이미지는 2/n^2 격자, 홀수면 오류남;;
pattern = np.zeros((round(height/2+0.1), round(width/2+0.1), ch, 4), dtype=np.uint8)
print(pattern.shape)

xturn=False
yturn=False
h=int(height/block)
w=int(width/block)
print(h, w)
try: 
    for y in range(block):
        y1=y*h
        y2=int(y1/2)-yturn*int(h/2)
        for x in range(block):
            x1=x*w
            x2=int(x1/2)-xturn*int(w/2)
            num=(yturn << 1) | int(xturn)
            print(",",f'({x1},{y1})({x1+w},{y1+h})|[{x2},{y2}][{x2+w},{y2+h}]', num, end=' ')
            pattern[y2:y2+h, x2:x2+w, :, num] = image[y1:y1+h, x1:x1+w]
            xturn=not xturn
        print()
        xturn=False
        yturn=not yturn

    cv2.imshow(f"origin", image)

    for i in range(4):
        cv2.imshow(f"Pattern{i}", cv2.resize(pattern[:,:,:,i], (width,height)))

    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
except Exception as e:
    print(f'\nError: {pattern[y2:y2+h, x2:x2+w, :, num].shape}\n{image[y:y+h, x:x+w].shape}')
```
